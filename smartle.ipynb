```python
with open('words.txt') as file:
    words = file.readlines()
    words = [word.strip() for word in words]
```

```python
print(words)
```

```python
alphabet_frequency = {}
for word in words:
    for letter in word:
        if letter in alphabet_frequency:
            alphabet_frequency[letter] += 1
        else:
            alphabet_frequency[letter] = 0
```

```python
import matplotlib.pyplot as plt

# Extract keys (letters) and values (frequencies)
af_list = list(alphabet_frequency.items())
af_list = sorted(af_list, key=lambda x: x[1])
letters = [x[0] for x in af_list]
frequencies = [x[1] for x in af_list]

# Create the bar chart
plt.bar(letters, frequencies)  # Letters on x-axis, frequencies on y-axis

# Customize the plot (optional)
plt.xlabel("Letter")
plt.ylabel("Frequency")
plt.title("Letter Frequency Distribution")

plt.savefig("letter_frequency_plot.png", dpi=300, bbox_inches='tight')  # Adjust filename and format

# Display the plot
plt.show()

```

```python
print(sorted(alphabet_frequency.items(), key=lambda x: x[1]))
```

```python
word_weight = {}
for word in words:
    weight = 0
    for letter in word:
        weight += alphabet_frequency[letter]
    word_weight[word] = weight
```

```python
word_weight = [t[0] for t in sorted(word_weight.items(), key=lambda x: x[1])]
```
